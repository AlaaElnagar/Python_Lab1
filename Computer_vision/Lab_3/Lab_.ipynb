```python
import cv2
import numpy as np 
import cv2 as cv


#---------------step1 get the main points using cursor 

#---------------step2 set the point in array 1 ,array2 

#---------------step3 geometric transform 
# mouse callback function


def get_pose(event,x,y,flags,param):

    if event == cv.EVENT_LBUTTONDOWN:
        print(x,y)

        
    
    
        
cv2.namedWindow('image')

img = cv2.imread('perspective02.jpg')
img = cv2.resize(img,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)

cols ,rows =rows,cols = img.shape[:2]

cv.setMouseCallback('image', get_pose)

#---------------------Transform section 
pts1 = np.float32(
    [
        [638,  217], [0,  217],
        [468, 167], [170, 167] ,
       # [582 ,341] , [701 ,341 ]
    
    ]
)

pts2 = np.float32(
    [
        [415,  339], [265,  339],
        [457, 0], [180, 4],
       
      
        #[582 ,341] , [701 ,341 ]
    ]
)

    

M = cv2.getPerspectiveTransform(pts1 ,pts2)

dst = cv2.warpPerspective(img, M, (cols,rows))

SHOW(dst)



```

```python
import cv2
import numpy as np 
import cv2 as cv



def get_pose(event,x,y,flags,param):

    if event == cv.EVENT_LBUTTONDOWN:
        print(x,y)

from PIL import Image

def SHOW(img):
    display(Image.fromarray(img[:,:,::-1]));        
    
    
        
cv2.namedWindow('image')

img = cv2.imread('perspective01.png')
img = cv2.resize(img,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)

cols ,rows =rows,cols = img.shape[:2]

cv.setMouseCallback('image', get_pose)

pts1 = np.float32(
    [
        [307,  79], [0,  79],
        [0, 124], [400, 124] ,
       # [582 ,341] , [701 ,341 ]
    
    ]
)

pts2 = np.float32(
    [
        [250,  79], [229,  79],
        [0, 269], [478, 269],
       
      
        #[582 ,341] , [701 ,341 ]
    ]
)

    

M = cv2.getPerspectiveTransform(pts1 ,pts2)

dst = cv2.warpPerspective(img, M, (cols,rows))

SHOW(dst)


```
