```python
import numpy as np
import cv2 as cv
import cv2 as cv2

from matplotlib import pyplot as plt
import time

def nothing(x):
    pass

cv.namedWindow('Output')
cv.createTrackbar('blockSize', 'Output', 0, 30, nothing)
cv.createTrackbar('apertureSize', 'Output', 0, 30, nothing)
cv.createTrackbar('k', 'Output', 0, 100, nothing)
cv.createTrackbar('dilation', 'Output', 0, 100, nothing)

#cv.cornerHarris(src_gray, blockSize, apertureSize, k)
filename = '001.png'
src=cv.imread(filename)
img = cv.imread(filename)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
gray = np.float32(gray)

while 1:
    #img = img.copy()
    img = cv.imread(filename)
    #gray=gray.copy()
    #--------------------------------------------------------calc Algorithm time -----------------------------------------------------------------------
    start_time = time.time()

    blockSize = cv2.getTrackbarPos('blockSize', 'Output')
    apertureSize = cv2.getTrackbarPos('apertureSize', 'Output')
    k = cv2.getTrackbarPos('k', 'Output')
    k=k/100
    dilation = cv2.getTrackbarPos('dilation', 'Output')
    dilation=dilation/100
    if k == 0:
        k=0.01
    if blockSize == 0:
        blockSize=2
    if (apertureSize%2 )!= 1:
        apertureSize+=1
    
    if apertureSize == 0:
        apertureSize=1
    if dilation ==0 :
        dilation=0.01

    dst = cv.cornerHarris(gray,2,apertureSize,k)
    #result is dilated for marking the corners, not important
    dst = cv.dilate(dst,None)
    # Threshold for an optimal value, it may vary depending on the image.
    img[dst>dilation*dst.max()]=[0,0,255]

    #--------------------------------------------------------End of calc Algorithm time 
    end_time = time.time()
    total_time = end_time - start_time

    cv.imshow('Output',img)
    if cv.waitKey(1) & 0xff == 27:
        break
       
cv.destroyAllWindows()
print("Time: ", total_time)


titles = ['The original image', 'harris corner ']
images = [src, img]

plt.figure(figsize=(30,15), facecolor=(1, 1, 1))
for i in range(2):
    plt.subplot(1, 2, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```
