```python
import tensorflow as tf
import numpy as np
import time
import random

x_list = list()
y_list = list()
z_list = list()

for i in range (10):
    x = int (random.randint(1, 100))
    y = int (random.randint(1, 100))
    z =int ( 5*x - 3*y + 7)
    x_list.append(x)
    y_list.append(y)
    z_list.append(z)

x_list= np.array(x_list,  dtype=float)
y_list= np.array(y_list,  dtype=float)
z_list= np.array(z_list,  dtype=float)
```

```python
t = time.time()
merged_array = np.stack([x_list, y_list], axis=1)

model0 = tf.keras.Sequential(
tf.keras.layers.Dense(units=1, input_shape=[2])   
)


model0.compile(
        loss='mean_squared_error',
        optimizer=tf.keras.optimizers.Adam(0.1)
    )


model0.fit(merged_array,z_list, epochs=500)

print (merged_array)
print (z_list)
```
